```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sentences = ['I played Ghost of Tsushima',
        'Call of Duty Vanguard is great',
        'Max payne is a classic game',
        'Far Cry 6 is hard to play',
        'Started playing God of War recently']
```

```python
voc_size = 10000
```

```python
onehot = [one_hot(words,voc_size)for words in sentences] 
print(onehot)
```

```python
sent_length = 8
embedded_docs = pad_sequences(onehot,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs))
```
